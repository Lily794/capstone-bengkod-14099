```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Height'] = pd.to_numeric(df['Height'], errors='coerce')
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')
df['FCVC'] = pd.to_numeric(df['FCVC'], errors='coerce')
df.info()
```

```python
print(df[['Age', 'Height', 'Weight']].describe())
```

```python
df[['Age', 'Height', 'Weight']].hist(bins=40, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Age', 'Height', 'Weight']])
plt.show()
```

```python
print("\nDistribusi Kelas pada Kolom Target 'NObayesdad':")
print(df['NObeyesdad'].value_counts())
```

```python
correlation_matrix = df[['Age', 'Height', 'Weight']].corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
